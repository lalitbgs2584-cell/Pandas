```python
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv("./data/raw_data.csv"))
```

```python
copy = df.copy()
copy
```

```python
copy["tax"] = copy["income"].apply(lambda x: "20%" if x>=50000 else "10%")
```

```python
copy
```

```python
gender_map = {"Male": "M","Female":"F","Unknown":"U"}
copy["gender"] = copy["gender"].map(gender_map)
copy
```

```python
copy = copy.assign(new_income = copy["income"]*1.1)
copy
```

```python
copy["country"] = copy["country"].replace("USA","US")
copy
```

```python
copy.rename(columns={"new_income":"new income"},inplace=True)
copy
```

```python
min_income = copy["income"].min()
min_income
```

```python
copy["Ranking"] = copy["income"].rank()
copy.rename(columns={"Ranking":"ranking"})
```
