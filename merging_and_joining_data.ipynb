```python
import pandas as pd
```

```python

df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Adam", "Bob", "Charlie", "Dave"]
})

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104, 105],
    "customer_id": [1, 2, 1, 3, 5],   # notice 5 does not exist in customers
    "amount": [250, 450, 300, 150, 500]
})
```

```python
df_customers
```

```python
df_orders
```

```python
pd.merge(df_customers,df_orders,on="customer_id")  # inner join
```

```python
left_join = pd.merge(
    df_customers,
    df_orders,
    on="customer_id",
    how="left"
)

left_join
```

```python
right_join = pd.merge(
    df_customers,
    df_orders,
    on="customer_id",
    how="right"
)

right_join
```

```python
outer_join = pd.merge(
    df_customers,
    df_orders,
    on="customer_id",
    how="outer"
)

outer_join
```
