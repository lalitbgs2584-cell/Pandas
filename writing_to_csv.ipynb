```python
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv("./data/raw_data.csv"))
copy = df.copy()
```

```python
copy = copy.drop_duplicates()
mean_age = copy["age"].mean()
mean_income = copy["income"].mean()
copy["age"] = copy["age"].fillna(mean_age)
copy["income"] = copy["income"].fillna(mean_income)
new_data = copy.dropna()
```

```python
new_data.to_csv("./data/new_data.csv")
```
