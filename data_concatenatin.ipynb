```python
import pandas as pd
```

```python

df1 = pd.DataFrame({
    "id": [1, 2],
    "name": ["Adam", "Bob"]
})

df2 = pd.DataFrame({
    "id": [3, 4],
    "name": ["Charlie", "Dave"]
})

row_concat = pd.concat([df1, df2])
column_concat = pd.concat([df1, df2],axis=1)
print(row_concat)
print(column_concat)
```
