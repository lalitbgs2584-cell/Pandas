```python
import pandas as pd
```

```python
data = {
    "country":["USA","USA","India","India"],
    "year":[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
melted_df = df.melt(
    id_vars=["country","year"],
    value_vars=["sales","profit"],
    var_name="metrics",
    value_name="value"
)
```

```python
pivot_df = melted_df.pivot(
    index=["country", "year"],
    columns="metrics",
    values="value"
)
pivot_df
```
