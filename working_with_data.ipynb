```python
import pandas as pd
```

```python
data = pd.read_csv("./data/employee_data.csv")
df = pd.DataFrame(data)
df
```

# DataFrame Method

```python
print(df.head())      
# Shows the first 5 rows of the DataFrame (default = 5 rows)
# Useful to quickly inspect the top of the dataset

print(df.tail())      
# Shows the last 5 rows of the DataFrame
# Helpful to check how the dataset ends

print(df.isnull())    
# Returns a DataFrame of True/False values
# True means the value is missing (NaN), False means value is present

print(df.describe())  
# Gives statistical summary of numerical columns
# Includes count, mean, std, min, 25%, 50%, 75%, max

print(df.columns)     
# Displays the column names of the DataFrame

print(df.info())      
# Provides summary information:
# - Number of entries
# - Column names
# - Data types
# - Non-null counts
# - Memory usage

print(df.sample())    
# Returns one random row from the DataFrame (default = 1 row)
# Useful for random inspection of data

print(df.nunique())   
# Shows the number of unique values in each column
# Helpful for understanding categorical features
```
